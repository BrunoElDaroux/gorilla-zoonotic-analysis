```python
# ── Configuration ─────────────────────────────────────────────────────────────
RANDOM_SEED  = 42          # For full reproducibility
START_YEAR   = 2015        # First field season
END_YEAR     = 2023        # Last field season
RAW_DATA_DIR = PROJECT_ROOT / 'data' / 'raw'
PROC_DATA_DIR = PROJECT_ROOT / 'data' / 'processed'
FIG_DIR      = PROJECT_ROOT / 'outputs' / 'figures'
RPT_DIR      = PROJECT_ROOT / 'outputs' / 'reports'

# Create all necessary directories
for d in [RAW_DATA_DIR, PROC_DATA_DIR, FIG_DIR, RPT_DIR]:
    d.mkdir(parents=True, exist_ok=True)
    print(f'  Directory ready: {d}')

print(f'\nSeed: {RANDOM_SEED} | Period: {START_YEAR}–{END_YEAR}')
```

## Cell 3 — Preview Simulation Parameters

```python
# ── Gorilla Groups Overview ───────────────────────────────────────────────────
print('=== GORILLA GROUPS CONFIGURED ===')
groups_df = pd.DataFrame(GORILLA_GROUPS).T.reset_index()
groups_df.columns = ['Group', 'Population Size', 'Home Range', 'Altitude (m)', 'Visit Frequency']
print(groups_df.to_string(index=False))

print(f'\nTotal gorilla population (approx): {sum(g["size"] for g in GORILLA_GROUPS.values())}')
print(f'Age categories: {AGE_CATEGORIES}')
print(f'\nStudy period: {START_YEAR}-01-01 to {END_YEAR}-12-31')
print(f'Total study days: {(datetime(END_YEAR, 12, 31) - datetime(START_YEAR, 1, 1)).days + 1}')
```

## Cell 4 — Run Data Generator

```python
# ── Initialize and Run Generator ─────────────────────────────────────────────
gen = GorillaDataGenerator(
    seed=RANDOM_SEED,
    start_year=START_YEAR,
    end_year=END_YEAR,
)

# Generate all datasets (this may take 30–90 seconds or longer depending on your machine)
datasets = gen.generate_all(output_dir=str(RAW_DATA_DIR) + '/')

# Unpack for easy access
demographics_df    = datasets['gorilla_demographics']
climate_df         = datasets['seasonal_climate_data']
visit_logs_df      = datasets['tourist_visit_logs']
tourist_health_df  = datasets['tourist_health_records']
gorilla_health_df  = datasets['gorilla_health_events']
```

## Cell 5 — Quick Dataset Validation

```python
# ── Dataset Validation ────────────────────────────────────────────────────────
print('\n=== DATASET SUMMARY ===')
print(f'{'Dataset':<30} {'Rows':>8} {'Columns':>10} {'Nulls':>8}')
print('-' * 60)

for name, df in datasets.items():
    null_pct = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100)
    print(f'{name:<30} {len(df):>8,} {df.shape[1]:>10} {null_pct:>7.1f}%')

print('\n=== REFERENTIAL INTEGRITY CHECKS ===')

# All gorilla IDs in health events exist in demographics
health_gorilla_ids = set(gorilla_health_df['gorilla_id'].unique())
demo_gorilla_ids   = set(demographics_df['gorilla_id'].unique())
orphan_ids = health_gorilla_ids - demo_gorilla_ids
print(f'Health event gorilla IDs in demographics: {len(orphan_ids) == 0} (orphans: {len(orphan_ids)})')

# All group IDs consistent
health_groups  = set(gorilla_health_df['group_id'].unique())
visit_groups   = set(visit_logs_df['group_id'].unique())
tourist_groups = set(tourist_health_df['group_visited'].unique())
expected_groups = set(GORILLA_GROUPS.keys())
print(f'Group IDs consistent across tables: {health_groups == visit_groups == tourist_groups == expected_groups}')
print(f'Groups: {sorted(health_groups)}')

# Date ranges
print(f'\nDate ranges:')
print(f'  Climate:        {climate_df["date"].min()} → {climate_df["date"].max()}')
print(f'  Tourist visits: {visit_logs_df["visit_date"].min()} → {visit_logs_df["visit_date"].max()}')
print(f'  Gorilla events: {gorilla_health_df["event_date"].min()} → {gorilla_health_df["event_date"].max()}')
```

## Cell 7 — High-Level Summary Statistics

```python
import sqlite3

# ── SQL summary using in-memory SQLite ────────────────────────────────────────
# This demonstrates SQL skills and builds the db used in later notebooks
conn = sqlite3.connect(':memory:')

# Load all tables
demographics_df.to_sql('gorilla_demographics', conn, index=False, if_exists='replace')
gorilla_health_df.to_sql('gorilla_health_events', conn, index=False, if_exists='replace')
tourist_health_df.to_sql('tourist_health_records', conn, index=False, if_exists='replace')
visit_logs_df.to_sql('tourist_visit_logs', conn, index=False, if_exists='replace')
climate_df.to_sql('seasonal_climate_data', conn, index=False, if_exists='replace')

print('Tables loaded into SQLite in-memory database.')

# Summary query: illness events per group per year
query = """
SELECT
    ghe.group_id,
    CAST(strftime('%Y', ghe.event_date) AS INTEGER) AS year,
    COUNT(*) AS total_events,
    SUM(CASE WHEN ghe.illness_type = 'respiratory' THEN 1 ELSE 0 END) AS respiratory_events,
    SUM(CASE WHEN ghe.transmission_source = 'tourism_linked' THEN 1 ELSE 0 END) AS tourism_linked,
    SUM(CASE WHEN ghe.severity = 'fatal' THEN 1 ELSE 0 END) AS fatal_events,
    ROUND(AVG(ghe.days_ill), 1) AS avg_days_ill
FROM gorilla_health_events ghe
GROUP BY ghe.group_id, year
ORDER BY ghe.group_id, year
"""

summary_sql = pd.read_sql(query, conn)
print('\n=== ILLNESS EVENTS PER GROUP PER YEAR (SQL) ===')
print(summary_sql.to_string(index=False))

conn.close()
```
